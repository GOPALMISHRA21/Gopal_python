```python
w = df[(df['duration'] =='3 Seasons') & (df['duration'] =='2 Seasons') & (df['duration'] =='3 Seasons') ].index
df.drop(w,inplace=True)

```

```python
def clean_duration_count(value):
    if isinstance(value,float):
        return value
    value_parts = value.split()
    duration_time = value_parts[0].replace(',','')
    return duration_time
df['duration'].apply(clean_duration_count) 
```

```python
duration=df['duration'].apply(clean_duration_count).apply(pd.Series)
duration
px.bar(duration.head(10),title='top 10 durations movies')
```
