# Numerical

```python
df['Population'].plot(kind='line',figsize=(5,5),color='red')
```

```python
df.set_index('City',inplace=True)
```

```python
df['Population'].plot(kind='line',figsize=(5,5),color='red')
```

```python
df["Population"].head(100).plot(kind='line',figsize=(20,5),style='ro-')
ticks=np.arange(0,100)
plt.xticks(ticks,df.index[ticks],rotation=90)
plt.show()
```

```python
df['Sexratio'].plot(kind='hist',figsize=(20,5))
```

```python
df["Metro_Population"].head(100).plot(kind='line',figsize=(20,5),style='ro--')
ticks=np.arange(0,100)
plt.xticks(ticks,df.index[ticks],rotation=90)
plt.show()
```

```python
df["Population"].head(100).plot(kind='line',figsize=(20,5),style='ro-',logy=True)
ticks=np.arange(0,100)
# draw a line
plt.axhline(y=10**6,color='b',linestyle='--')
plt.xticks(ticks,df.index[ticks],rotation=90)
plt.show()
```

```python
df[["Population",'Metro_Population']].plot(kind='line',figsize=(20,6),logy=True)
```

```python
df['Pop_Diff']=df['Metro_Population']- df['Population']
df.head()
```

```python
df[['Population','Pop_Diff']].plot(kind='line',figsize=(20,6))
```
