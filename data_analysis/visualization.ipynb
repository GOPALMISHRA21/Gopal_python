a simple graph with sine wave and cosine wave

```python
t=np.linspace(0,10,100)
flg = plt.figure(figsize=(20,5))
plt.plot(t,np.sin(t),'r-',label='sin(t)')
plt.plot(t,np.cos(t),'b--',label='cos(t)')
plt.legend()
plt.show()
```

A figure is the outer  most container foa Matploylib plot.The Figure 

```python
fig,ax = plt.subplots(nrows=1,ncols=2,figsize=(20,6))
ax[0].plot(t,np.sin(t),'r-',label='sin wave of t')
ax[1].plot(t,np.cos(t),'b--',label='cosine wave of t')
ax[0].legend()
ax[1].legend()
plt.show
```

## adjusting the labels, ticks,legends and x-axis,y-axis

```python
x=np.linspace(0,10,100)
plt.plot(x,np.sin(x),'b',label='sine wave of x')
plt.xlabel('x values',fontsize=14,fontdict={'color':'green'})
plt.ylabel('y values',fontsize=14,fontdict={'color':'green'})
plt.title('Sine wave',fontsize=16,fontdict={'color':'green','weight':'bold'})
plt.show()
```

```python
x=np.linspace(0,10,100)
plt.plot(x,np.sin(x),'b',label='sine wave of x')
plt.xlabel('x values',fontsize=14,fontdict={'color':'green'})
plt.ylabel('y values',fontsize=14,fontdict={'color':'green'})
plt.title('Sine wave',fontsize=16,fontdict={'color':'green','weight':'bold'})
plt.text(4,.5,'y=sin(x)',fontsize=10,fontdict={'color':'red',})
plt.text(1,.25,'crest',fontsize=10,fontdict={'color':'red',})
plt.text(4,-.5,'trough',fontsize=10,fontdict={'color':'red',})
#annotate

plt.annotate('local max',xy=(1.6,1),xytext=(1,1.5),
            arrowprops={'facecolor':'green'})

plt.annotate('local max',xy=(5,-1),xytext=(8,-.75),
            arrowprops={'facecolor':'red'})
plt.show()
```

```python
# changing the background colour
fig=plt.figure(figsize=(20,6),facecolor='lightblue')
x=np.linspace(0,10,100)
plt.plot(x,np.sin(x),'b',label='sine wave of t')
# change axes background color
ax=plt.gca()  # get current graph 'axes'

ax.set_facecolor('#8943ff')
plt.show()
```

```python
# scatter plot
x=np.random.randn(100)
y=np.random.randn(100)
plt.scatter(x,y,marker='o',color='red')
plt.show()
```

```python
# bubble plot
fig=plt.figure(figsize=(20,6))
x=np.random.randn(100)
y=np.random.randn(100)
size=np.random.randint(10,500,100)
plt.scatter(x,y,marker='o',c=size,s=size,alpha=.5,edgecolors='black',cmap='viridis')
plt.colorbar()
plt.show()
```

```python
# pie chart

gender_count=np.array([120,45])
gender_labels=np.array(['Female','Male'])
# text properties

props={'color':'white','weight':'bold','fontsize':14}
plt.pie(gender_count,labels=gender_labels,textprops=props)
plt.show()
```

```python
plt.pie(gender_count,labels=gender_labels,textprops=props,autopct='%.1f%%')  # autopct
plt.show()
```

```python
continents=['Africa','Asia','Europe','North America','South America','Ocenia']
population=[126,4436,738,579,422,38]
props={'color':'white','weight':'bold','fontsize':8}
plt.pie(population,labels=continents,autopct='%.1f%%',textprops=props)
plt.show()
```

```python
# donut chart
fig=plt.figure(figsize=(10,5),facecolor='black')
continents=['Africa','Asia','Europe','North America','South America','Ocenia']
population=[126,4436,738,579,422,38]
tprops={'color':'yellow','weight':'bold','fontsize':8}
wprops={'width':0.6,'linewidth':2}   # wedgeprops
plt.pie(population,labels=continents,autopct='%.1f%%',textprops=tprops,wedgeprops=wprops)
plt.show()
```

```python
# Bar Plot
sales=np.array([100,200,300,400,500,420,300,500,600])
months=np.array(['Jan','Feb','Mar','Apr','May','Jun','July','Aug','Sep'])
plt.bar(months,sales)
plt.show()
```

```python
plt.barh(y=months,width=sales)
plt.show()
```

```python
plt.bar(x=months,height=sales,color='green',edgecolor='black',alpha=.5)
```

```python
colors=['red','green','blue','yellow','orange','purple','brown','pink','grey']
plt.bar(x=months,height=sales,color=colors,edgecolor='black',alpha=.5)
```

```python
colors=['red','green','blue','yellow','orange','purple','brown','pink','grey']
plt.bar(x=months,height=sales,color=colors,edgecolor='black',alpha=.5,hatch='/')
```

```python
# bar chart with text
plt.bar(x=months,height=sales)
for i in range(len(sales)):
    plt.text(x=i,y=10,s=sales[i],ha='center',fontdict={'color':'white'})
plt.show()
```

```python

# visualization
# stem plot
fig= plt.figure(figsize=(20,6))
x= np.linspace(0, 10, 50)
y= np.sin(x)
plt.stem(x, y, linefmt='r:', markerfmt='r*',basefmt='g')
plt.show()
```

```python
#hexbin plot
fig = plt.figure(figsize=(8,6))
x = np.random.randn(1000)
y = np.random.randn(1000)
plt.hexbin(x, y, gridsize= 25, cmap='viridis_r')
plt.colorbar()
plt.show()
```

```python
# box plot
fig=plt.figure(figsize=(20,6))
x=np.random.randint(1,10000,(100,3))
plt.boxplot(x,meanline=True,showmeans=True,showfliers=True,notch=True)
plt.show()
```
